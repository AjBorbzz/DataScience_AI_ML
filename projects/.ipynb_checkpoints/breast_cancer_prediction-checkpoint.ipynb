# Label Encoding

```python
#Get Datatypes of each column in our dataset
df.dtypes
```

```python
#Encode the diagnosis values
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['diagnosis'] = label_encoder.fit_transform(df['diagnosis'].values)
```

```python
#display df
df
```

# Split Dataset & Feature Scaling

```python
#Splitting the dataset into independent and dependent datasets 
X = df.iloc[:,2:].values
y = df['diagnosis'].values
```

```python
#Splitting datasets into training(75%) and testing(25%)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25, random_state=5)
```

```python
#Scaling the data(feature scaling)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
#print data
X_train
```

# Build a Logistic Regression Model

```python
#build a logistic regression classifier
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
#make use of trained model to make predictions on test data
predictions = classifier.predict(X_test)
```

### Let us evaluate the performance of the model by using the confusion matrix
|     |     |
| --- | --- |
| TP  | FP  |
| FN  | TN  |

```python
#plot confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predictions)
sns.heatmap(cm, annot=True)
cm
```

```python
#get accuracy score for model
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, predictions)
acc
```

#### Conclusion
In this machine learning project, we aimed to build a classification model to predict breast cancer outcomes based on various features. After training the model and evaluating its performance on the test set, we obtained insightful results.

#### Confusion Matrix Analysis
The confusion matrix provides a detailed breakdown of the model's predictions compared to the actual outcomes. From the confusion matrix:

* True Positives (TP): 87
* True Negatives (TN): 51

* False Negatives (FN): 4

We observe that the model has performed remarkably well in correctly classifying instances of both malignant and benign tumors. With only 1 false positive and 4 false negatives, the model demonstrates a high level of precision and recall.
